# CIC-IDS2017 Dataset Inspection

This notebook performs an initial exploratory analysis of the CIC-IDS2017 dataset.

The goal is to:
- inspect dataset structure
- check for missing values
- detect infinite values
- analyze class distribution


## Import packages


```python
import pandas as pd
import numpy as np
from pathlib import Path
```

## Dataset paths

```python
DATA_DIR = Path('C:\\Users\\emilk\\pr_lic\\data\\raw')
file_names = [f for f in DATA_DIR.glob('*.csv')]

print("Number of files: ", len(file_names))
print("Files: ")
for f in file_names:
    print(f)
```

## Load sample data

```python
file = file_names[3]

data = pd.read_csv(file)

df = pd.DataFrame(data)
df.columns = df.columns.str.strip()

print("File:", file.name)
print("Shape:", df.shape)

df.head()
```

## Dataset structure

```python
df.info()
```

## Duplicate records

```python
duplicates = df.duplicated().sum()

print("Number of duplicate rows:", duplicates)
```

## Missing values

```python
missing = df.isna().sum()

print(missing[missing>0])
```

## Infinite values

```python
numeric_df = df.select_dtypes(include=[np.number])

infinite = np.isinf(numeric_df).sum()

print(infinite[infinite>0])
```

## Class distribution

```python
df['Label'].value_counts()
```
